```python
# import the required libraries
import sweetviz as sv
import pandas as pd

print("SweetViz Version : {}".format(sv.__version__))
```

# Load Dataset

```python
data = pd.read_csv('/content/drive/MyDrive/ProjectsData/spotify-2023.csv', encoding='latin-1')
data.head()
```

# Create the report
for creating the report we have 3 functions:

analyze() for a single dataset
compare() for compare 2 datasets (e.g. Test versus Train)
compare_intra() for comparing 2 sub-populations within the same dataset

```python
report = sv.analyze(data)
```

# Generate output
Once we have created our report object, then simply pass it to one of the two ‘show’ functions:


*   show_html() function will create & save an HTML report (HTML page) at the given file path, &

*   show_notebook() function will embed the report inside a Notebook.






```python
report.show_html('Report.html')
```

```python
report.show_notebook()
```
